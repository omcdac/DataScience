**Operators**


There are operators present in the pythons,

1. Arithmatic
2. logical
3. Relational
4. Assignments

```python
# 1. Arithmatic
num1 = 4; num2 = 2;
add = num1 + num2 ; print(add); # Addition
sub = num1 - num2 ; print(sub);  # Substraction
mul = num1 * num2 ; print(mul); # Multiplication
div = num1 / num2 ; print(div); # Division
floor = num1 // num2 ; print(floor); # floor division
modulo = num1 % num2 ; print(modulo);  # modulo division -- It is reminder of the number
power = num1 ** num2 ; print(power); # Power of
```

```python
#2. Assginment Operators
num1 = 66
num1 //= 2
print (num1)
```

```python
#3. Relational Operators
num1 = 5
num2 = 2
print("Greater than:", num1 > num2)
print("Greater than or equal to:", num1 >= num2)
print("Less than:", num1 < num2)
print("Greater than or equal to:", num1 <= num2)
print("Equal to:", num1 == num2)
print("Not equal to:", num1 != num2)
```

```python
#4 Logical Operators
num1 = 5
num2 = 2
print("AND:", (num1 > num2) and (num1 < 10) )
print("OR:", (num1 > num2) or (num1 < 1) )
```
