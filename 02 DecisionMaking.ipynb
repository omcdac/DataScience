**Decision Making**

```python
# Check whether number is Positive or Negative
# It is important to place sentences properly with correct indentation...
# It uses relational or logical operators to decide the condition
num = int(input("Enter the number "))
if num > 0:  #It is compulsory, Condtion, Non-repeatable
  print("Number is Positive")
  print("OK")
else:   #Optional, No condition, Non-repetable
  print("Number is Negative")
  print("NOT OK")
```

```python
#Nested conditions
num = int(input("Enter the number "))
if num > 0:
  print("Number is Positive")
elif num == 0:  #Optinal, Can be written multiple times (repeatable)
  print("It is ZERO")
else:
  print("Number is Negative")
```

```python
# Write a program to check the number is odd or even

num = int(input("Enter a number "))

if num == 0:
  print("The number is ZERO")
elif num % 2 == 0:
  print("The number is EVEN")
else:
  print("The number is ODD")
```

**Loop Control statements**

```python
#1. While, For Loops (Nested Also posible)

count = 11  # Initialize
while count <= 20: # Condition
  print(count) # Operations
  count +=1
```

```python
## Adition of first 10 odd numbers
add = 0
num = 1
while num <= 20:
  add = add + num
  num += 2

print(add)
```

**For Loop**

```python
# For loops : Iterates through the collection of values/elements
# It has predefined number of iterations

for i in 10,20,30,40:
  print(i*i)
```

```python
# For loops
num = [10,20,30,40]
for i in num:
  print(i*i)
```

```python
# prompt: addition of first 10 odd numbers

add = 0
num = 1
while num <= 19:
  add = add + num
  num += 2
print(add)
```
