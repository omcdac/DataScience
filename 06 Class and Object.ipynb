**Object Oriented Programming in Python**

Class - (skeleton/blue print)
  1. Attributes
  2. Methods

Object -
  Exact value

1. Instance means Attributes
2. Fuction means Action

# Define the class

```python
#Define the class
class Person:   #Class name starts with Capital Letter
  # constructor
  def __init__(self, n, a):
    self.name = n # these are instance variables, i.e seperate copies are created for every object
    self.age = a   # name and age are attributed /properties
  def show(self):
    print("My data:", self.name, self.age)
  def isEligibleForVoting(self):
    if self.age >= 18:
      print("Eligible for voting")
      return True
    else:
      print("Not eligible for voting")
      return False


ob1 = Person("Om", 28)
ob2 = Person("Ram", 30)

ob1.name, ob1.age
ob2.name, ob2.age

ob1.show()
ob2.show()

ob2.isEligibleForVoting()
ob1.isEligibleForVoting()
```
