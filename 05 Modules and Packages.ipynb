```python
#1. import keyword
#2. import as
#3. from import

# Important link: https://docs.python.org/3/py-modindex.html
# Use external modules/functions into your program
# these are used to read e.g

# Method-1: Import keyword

import math
math.sqrt(16)
math.pi
#help(math)
```

```python
import time
time.sleep(5)  # add delay
time.asctime() # system time

#
```

```python
#Method-2: Aliasing the functions
import math as m # alias of math as m - renaming - for module reference
m.sqrt(16), m.pow(2,3), m.log10(100)
```

```python
#Method-3: Selective import

from math import sqrt, pow, log10, pi # Select which you require and then you can directly use them. #What can be imported - class, function, variable etc.

sqrt(16), pow(2,3), log10(100), pi
```
