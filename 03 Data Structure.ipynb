```python
#List comprehension: In a single line, create new list from original list
num = [10,20,30,40,50]
num_new = [n*n for n in num] # Such kind of operations is called as list comprehension.
num_new
```

**2. Tuple**

1. Tuples are imutable i.e Once tuple is created, the data can not be changed. The operation, which changes the content are not permitted in the tuple

```python
# E.g
num = (10,23,31,30,40) # Parenthesis are used to store the data, hence are tuple
type(num)
# Access the content of the tuple. Accessing is same as the list.
#num[3] = 56 # Not possible in the tuple, as it is trying to modify the content
# Other operations can be performed with tuple.
num = 10,23,31,30,40   # If no seperator is given, it is by default treated as tuple.
type(num)
```

```python
# Tuple comes with () bracket
num = (10,20,30,30,40)
type(num)
```

**Set**

1. set: the collection of unique and distinct (non repeated) values
2. Set is not indexable.
3. It is mutable
4. It is iterable

Operations:
Unions
Intersection
Differentiation

```python
#Set comes with {} bracket
num = {10,20,30,30,40, 40}
type(num)
num
```

```python
#Set comes with {} bracket
num = {10,20,30,30,40, 40}
num
#To add elements inside the set
num.add(5)
num
```

```python
#Set comes with {} bracket
set1 = {10,20,30,30,40, 40}
set2 = {10,20,50,60,70}
#set1.union(set2) # Unique values from both sets is taken
#set1.intersection(set2) # Common values from both sets is taken
#set1.difference(set2) # Values from set1 which are not in set2 is taken
set2.difference(set1) # Values from set2 which are not in set1 is taken
```

**4. Dictionary**

It is collection of key value pairs.

E.g is look-up table.



```python
# Dictionary comes with {} with data seperators ;
d = {'a':10, 'b':20, 'c':30}  # a,b,c are the keys (keys are uniques, same as indexing but not in order i.e they may be random in nature) and 10,20,30 are the values.
type(d)
d['a']  # We use keys to access the corresponding values.
# Replacing the value
d['b'] = 100 #If it is present, it will replace, other-wise it will add.
d['d'] = 400
d
```
