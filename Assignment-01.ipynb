Q-1: Write a program to read Celsius temperature and print equivalent Fahrenheit temperate on screen

```python
#Write a program to read Celsius temperature and print equivalent Fahrenheit temperate on screen
celsius = float(input("Enter temperature in celsius: "))
fahrenheit = (celsius * 9/5) + 32
print('%.2f Celsius is: %0.2f Fahrenheit' %(celsius, fahrenheit))
```

Q-2: Read radius of the circle from user and find the area and perimeter of it.

```python
#Read radius of the circle from user and find the area and perimeter of it.
import math
radius = float(input("Enter the radius of the circle: "))
area = math.pi * radius * radius
perimeter = 2 * math.pi * radius
print("Area of the circle is: %.2f" %area)
print("Perimeter of the circle is: %.2f" %perimeter)
```

Q-3: Read the amount and percentage of interest from the keyboard and find final amount after adding interest in original amount.

```python
# Read the amount and percentage of interest from the keyboard and find final amount after adding interest in original amount.
amount = float(input("Enter the principal amount: "))
interest = float(input("Enter the rate of interest: "))
years = float(input("Enter the number of years: "))
final_amount = amount * (1 + (interest / 100)) ** years
print("Final amount after adding interest is: %.2f" %final_amount)
```

Q6: Write a program to input basic salary and find output of gross salary using following:

DA = 75% of basic
HRA = 20% of basic

Conditions:
Basic < 10000 : gross = da + basic
Basic >= 10000 and < 20000 : gross = da + basic + 50% of hra
Basic >= 20000 : gross = basic + da + hra

Sample: Input and Output
5000 : 3750 + 5000 = 8750
12000 : 9000 + 12000 + 1200 = 22200
24000 : 18000 + 24000 + 4800 = 46800

```python
# .
basic_salary = float(input("Enter the basic salary: "))

da = 0.75 * basic_salary
hra = 0.20 * basic_salary

if basic_salary < 10000:
    gross_salary = da + basic_salary
elif basic_salary >= 10000 and basic_salary < 20000:
    gross_salary = da + basic_salary + 0.50 * hra
else:
    gross_salary = basic_salary + da + hra

print(f"Gross Salary: {gross_salary:.2f}")
```

Q7: Write a program to read a number from user and find the factorial of the number.

```python
# Write a program to read a number from user and find the factorial of the number.
number = int(input("Enter a number: "))
factorial = 1
for i in range(1, number + 1):
    factorial *= i
print(f"The factorial of {number} is {factorial}")
```

Q8: Write a program to read num1 and num2 from keyboard and find addition of all number between that range.

```python
# Write a program to read num1 and num2 from keyboard and find addition of all number between that range.
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
sum = 0
for i in range(num1, num2 + 1):
    sum += i
print("The sum of all numbers between", num1, "and", num2, "is", sum)
```

Q9: See the tuple given: num = (56,2,35,41,43,48,32,56,71,55,68)
Write a program to create two lists from it where first list contains only even numbers and second will contain only odd numbers.

```python
# See the tuple given: num = (56,2,35,41,43,48,32,56,71,55,68). Write a program to create two lists from it where first list contains only even numbers and second will contain only odd numbers.
num = (56,2,35,41,43,48,32,56,71,55,68)
even_list = []
odd_list = []
for i in num:
    if i % 2 == 0:
        even_list.append(i)
    else:
        odd_list.append(i)

print("Even list:", even_list)
print("Odd list:", odd_list)
```

Q10: Write a program to read a number from user. Create and use a user defined function to check whether it is odd or even?

```python
# Write a program to read a number from user. Create and use a user defined function to check whether it is odd or even?
def check_odd_even(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

number = int(input("Enter a number: "))
result = check_odd_even(number)
print(f"{number} is {result}")
```

Q11: Write a program to read a number from user and find the square root, sin, tan, log and factorial of this number using math module.

```python
# Write a program to read a number from user and find the square root, sin, tan, log and factorial of this number using math module.
import math
number = float(input("Enter a number: "))
square_root = math.sqrt(number)
sin = math.sin(number)
tan = math.tan(number)
log = math.log(number)
factorial = math.factorial(int(number))
print("Square root of", number, "is", square_root)
print("Sine of", number, "is", sin)
print("Tangent of", number, "is", tan)
print("Logarithm of", number, "is", log)
print("Factorial of", number, "is", factorial)
```

Q12: Write a Python class named ‘Rectangle’ with two attributes length and width and a method named ‘calculate’ which will compute the area of a rectangle. Create an object to pass length and width and call the method to calculate the area of rectangle.

```python
# Write a Python class named ‘Rectangle’ with two attributes length and width and a method named ‘calculate’ which will compute the area of a rectangle. Create an object to pass length and width and call the method to calculate the area of rectangle.
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def calculate(self):
        return self.length * self.width

length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
rectangle = Rectangle(length, width)
area = rectangle.calculate()
print("The area of the rectangle is:", area)
```
