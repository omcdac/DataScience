**Fuction**

Function is a block of code which only runs when it is called.

1. Built in functions (prefined)
2. User - defined functions

Functions can be re-defined


```python
#Fuction Definition. It is also, self contained block of statements
def show():
  print("Hello world")
  print("Welcome to python")

#Function call
show()
```

**Function Parameters**

```python
# Parameters to pass in the function
#Define the function
def square(n):
  sq = n*n
  print(sq)

# call the function
square(25)
```

**Returning the Value**

```python
# Function can return the value using keyword return
def compare(a,b):
  if a>b:
    return a
  else:
    return b

t = 5*compare(10,20)  # returned value we can use it anyware...
print(t)
```

```python
# Python, can return more thna one value at a time. This is unique feature of python
def calc(num):
  sq = num*num
  cu = num*num*num
  return sq,cu

#print("The Square and Cube are:", calc(5)) # One can call functions in print statement
x,y = calc(5)
print("Square and Cube are:",x,y)
```

**Default Arguments for the functions**

```python
# Default values to the functions. User can mention it
def add(x=5, y=7, z=3):
  a = x + y + z
  print(a)


#print('Call-1', add(4,5,6))
#print('Call-2', add(4,5)) # In this case default value of z as 3 has been taken.
#print('Call-3', add(4))
print('Call-4', add(y=4)) # If one want to change the value of given variable, then one can assign something like this
```
